## 关系图绘制

### `replot`  
用于绘制两个变量之间的关系图  

```python  
seaborn.relplot(x=None, y=None, hue=None, size=None, style=None, data=None, row=None, col=None, col_wrap=None, row_order=None, col_order=None, palette=None, hue_order=None, hue_norm=None, sizes=None, size_order=None, size_norm=None, markers=None, dashes=None, style_order=None, legend='brief', kind='scatter', height=5, aspect=1, facet_kws=None, **kwargs)
```

|属性|值|说明|
|----|----|---|
|hue|name in data|用于分组数据-生成不同颜色|
|size|name in data|用于分组数据-生成不同size|
|style|..|用于分组数据-生成不同样式|
|data|dataframe|数据源|
|row,col|name of variables|确定网格切面|
|col_wrap|int|wap the column variabe at this width|
|row_order,col_order|string list|Order to organize the rows and/or columns of the grid in,|
|palette| palette name, list, or dict|hue中使用的不同颜色|
|hue_order|list|确定显示顺序|
|hue_norm|tuple|归一化colormap|
|sizes|list,dict,tuple|size中的size大小|
|size_order| list|Specified order for appearance of the size variable levels|
|size_norm|...|...|
|markers|boolean, list, or dictionary|Object determining how to draw the markers for different levels of the style variable|
|legend|"breief"<br>"full"<br>False|numeric hue and size variables will be represented with a sample of evenly spaced values<br>every group will get an entry in the legend.|
|kind|scatter, line|你要画那种形式的|
|height|scalar|facet的高度|
|aspect|scalar|Aspect ratio of each facet|
|facet_kws|dict|同时设置FacetGrid|
|kwargs|||  

建议： 使用hue+size+style

* **Facet on the columns and rows:**

````python
tips = sns.load_dataset("tips")
tips.head()
````

* 创建二维关系图

````python
sns.relplot(x="total_bill", y="tip", data=tips)
````

* 创建四维关系图

````python
g = sns.relplot(x="total_bill", y="tip", hue="day",col="time", row="sex", data=tips)
````

* **个性化replot**

````python
g = sns.relplot(x="total_bill", y="tip", hue="time",size = "size", palette=['b','r'], sizes=(10,100),col="time", data=tips)
````

### scatterplot

```python  
seaborn.scatterplot(x=None, y=None, hue=None, style=None, size=None, data=None, palette=None, hue_order=None, hue_norm=None, sizes=None, size_order=None, size_norm=None, markers=True, style_order=None, x_bins=None, y_bins=None, units=None, estimator=None, ci=95, n_boot=1000, alpha='auto', x_jitter=None, y_jitter=None, legend='brief', ax=None, **kwargs)  

```

````python
sns.scatterplot(x="total_bill", y="tip", data=tips)
````

````python
sns.scatterplot(x="total_bill", y="tip", hue="time",style="time",data=tips)
````

````python
sns.scatterplot(x="total_bill", y="tip", hue="day",style="time",data=tips)
````

````python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
# sizes中定义最大最小size大小
sns.scatterplot(x="total_bill", y="tip", hue="size",size="size",sizes=(20, 200),palette=cmap, data=tips) 
````

````python
sns.scatterplot(x="total_bill", y="tip", hue="size",size="size",sizes=(20, 200), hue_norm=(0, 7), data=tips) 
````

````python
 markers = {"Lunch": "s", "Dinner": "X"}
sns.scatterplot(x="total_bill", y="tip", style="time",markers=markers, data=tips) 
````
